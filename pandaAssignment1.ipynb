```python
import pandas as pd
import numpy as np
```

```python
import glob
```

```python
# Loading CSV file
```

```python
all_files = glob.glob("./*.csv")
li = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
    
us_census = pd.concat(li, axis=0, ignore_index=True)
us_census.head(10)
```

```python
# Checking the data type of file.
```

```python
us_census.dtypes
```

```python

# Droping the unnamed file 
```

```python
us_census.drop(columns = ['Unnamed: 0'], inplace=True)
```

```python
us_census.dtypes
```

```python
# Seperating Male and Female Column
```

```python
us_census[['Male', 'Female']] = us_census.GenderPop.str.split("_", expand=True)
```

```python
us_census
```

```python
# Droping Gender Pop colunm 
```

```python
us_census.drop(columns = ['GenderPop'], inplace=True)
```

```python
us_census.head()
```

```python
# Deleting $ sign from the Income column
```

```python
us_census ['Income'] = us_census.Income.str.strip('$')
```

```python
us_census.head()
```

```python
# Replacing percentage (%) from all columns
```

```python
us_census = us_census.replace ('%', '', regex=True)
```

```python
us_census.head()
```

```python
# Replacing M from the male column
```

```python
us_census = us_census.replace ('M', '', regex=True)
```

```python
# Deleting F form the column
```

```python
us_census = us_census.replace ('F', '', regex=True)
```

```python
us_census.head()
```

```python
us_census.dtypes
```

```python
# Rounding the value of columns 
```

```python
us_census.loc[:,'Hispanic':'Income'] = round(us_census.loc[:,'Hispanic':'Income'].apply(pd.to_numeric),2)
```

```python
us_census.head()
```

```python
us_census.dtypes
```

```python
# To change the data type of Male column
```

```python
us_census['Male'] = us_census['Male'].astype(int)
```

```python
us_census
```

```python
us_census.drop(columns = ['Female'], inplace=True)
```

```python
us_census['Female'] = us_census['TotalPop'] - us_census['Male']
```

```python
us_census.head()
```

```python
# showing data in scatter plot
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(us_census['Female'], us_census['Income'])
plt.xlabel('Salary $', fontsize=18)
plt.ylabel('Female_Population', fontsize=16)
plt.show()
```

```python
# Checking any duplicate value exist
```

```python
us_census.duplicated().any()
```

```python
# delete duplicate values
```

```python
us_census.drop_duplicates(inplace=True)
```

```python
plt.scatter(us_census['Female'], us_census['Income'])
plt.xlabel('Salary $', fontsize=18)
plt.ylabel('Female_Population', fontsize=16)
plt.show()
```
